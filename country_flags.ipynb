Toy task. 
Realize CNN based model that gets flag of certain country and outputs similar flags

Dataset can be found on kaggle
https://www.kaggle.com/frabbisw/flags-dataset

Approach:
Lets take VGG16 Neural Net withoung dense layers and use it as image descriptor

Use knn with cosine distance to range vectorized flags

```python
import numpy as np
import keras
import tensorflow as tf
from keras.applications import VGG19,VGG16,ResNet50,InceptionV3
from keras.engine import Model
import os
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input as preprocess_input_vgg16
```

```python
# help function

def listdir_full(d):
    return [os.path.join(d, f) for f in os.listdir(d)]
```

```python
# download vgg16 without top layers

model = VGG16(weights='imagenet', include_top=False)
```

```python
# directory path

path='/Users/user/Data_science/Sberbank_test/flags-dataset/1'
all_flags = listdir_full(path)
```

```python
# preprocess images for vgg16

flags=[]
for path in all_flags:
    img = image.load_img(path, target_size=(224, 224)) 
    x = image.img_to_array(img)  
    x = np.expand_dims(x, axis=0)  
    x = preprocess_input_vgg16(x) 
    flags.append(x)
    
```

```python
# vectorizing images

vectorized=[]
for i in flags:
    vectorized.append(model.predict(i).ravel())
```

```python
# "fit" knn

knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(vectorized)
```

```python
# flags that look like Slovakia flag

vec = vectorized[158].reshape(1,-1)
dist, indices = knn.kneighbors(vec, n_neighbors=10)
similar_images = [(all_flags[indices[0][i]], dist[0][i]) for i in range(len(indices[0]))]

images=[]
for i in similar_images: 
    img = image.load_img(i[0], target_size=(224, 224))
    images.append(img)

plt.imshow(images[0])
fig=plt.figure(figsize=(20, 20))
columns = 5
rows = 1
for i in range(1, columns*rows +1):
    img = images[i]
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```
